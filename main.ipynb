```python
import numpy as np
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv(r"Summer-Olympic-medals-1976-to-2008.csv", encoding = 'latin1')
df
```

```python
# Display basic info
df.info()
df.head()
```

```python
# Check for missing value
print("Missing value:\n", df.isnull().sum())
```

```python
# Drop missing value if necessary
df.dropna(inplace=True)
print("Missing value:\n", df.isnull().sum())
```

```python
df['Year'] = df['Year'].astype(int)
df
```

```python
# Check countries name with '*'
print(df['Country'][df['Country'].str.contains('\\*', regex=True)].unique())
```

```python
# Count countries name with "*"
print("Total Countries name with '*':", df['Country'].str.contains('\\*', regex=True).sum())
```

```python
# Remove '*' from Country names
df['Country'] = df['Country'].str.rstrip('*')
```

```python
# Count countries name with "*"
print("Total Countries with *:", df['Country'].str.contains('\\*', regex=True).sum())
```

```python
# Filter for a specific country
country_name = df[df['Country'] == 'Bermuda']['Country'].values[0]
print(country_name)
```

```python
# Display basic info
df.info()
```

```python
# Unique sports and events
print("Total Unique City:", df['City'].nunique())
print("Total Unique Year:", df['Year'].nunique())
print("Total Unique Sport:", df['Sport'].nunique())
print("Total Unique Discipline:", df['Discipline'].nunique())
print("Total Unique Event:", df['Event'].nunique())
print("Total Unique Athlete:", df['Athlete'].nunique())
print("Total Unique Gender:", df['Gender'].nunique())
print("Total Unique Country_Code:", df['Country_Code'].nunique())
print("Total Unique Country:", df['Country'].nunique())
print("Total Unique Event_gender:", df['Event_gender'].nunique())
print("Total Unique Medal:", df['Medal'].nunique())
```

```python
# Check Mapping Between Country & Country_Code
# Find if a country has multiple codes
country_mapping = df.groupby('Country')['Country_Code'].nunique()
print(country_mapping[country_mapping > 1])  # Only show countries with >1 unique codes
```

```python
# Check Different code
print(df[df['Country'] == 'Serbia'][['Country', 'Country_Code']].drop_duplicates())
```

```python
# Change Serbia code
df.loc[df['Country'] == 'Serbia', 'Country_Code'] = 'SRB'  # Replace all Serbia codes with 'SRB'
```

```python
print(df[df['Country'] == 'Serbia'][['Country', 'Country_Code']].drop_duplicates())
```

```python
print("Total Unique Country_Code:", df['Country_Code'].nunique())
print("Total Unique Country:", df['Country'].nunique())
```

```python
# Export Cleaned Olmpic Data file
df.to_csv('Olympic_cleaned_data.csv', index=False, encoding='utf-8')
```
